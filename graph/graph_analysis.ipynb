```python
# Define source as attack entry point

source = "HMI"
path = nx.single_source_shortest_path(G, source)

print(path)
```

### 2) Root-casue analysis

```python
# Define target as attack target point

target = "PCV02"
path = nx.single_target_shortest_path(G, target)

print(path)
```

* Notes: The paths include shortest path to target from all nodes that reach target.
